```python
import os
import cv2
import numpy as np
```

```python
data =[]
labels =[]
```

```python
dataset_path = "dataset"
```

```python
for img in os.listdir(dataset_path + "/real"):
     path = os.path.join(dataset_path + "/real", img)
     image = cv2.imread(path)
     image = cv2.resize(image,(128,128))
     data.append(image)
     labels.append(0)
```

```python
for img in os.listdir(dataset_path + "/fake"):
     path = os.path.join(dataset_path + "/fake", img)
     image = cv2.imread(path)
     image = cv2.resize(image,(128,128))
     data.append(image)
     labels.append(1)
```
